```python
data = np.random.multivariate_normal(size=100, mean= [-5, 1], cov=[[3, 1],[1,1]])
X, Y = data[:, 0], data[:, 1]
```

```python
plt.scatter(X, Y)
plt.title ("Data") ; plt.xlabel("X axis") ; plt.ylabel("Y axis")
```

```python
# Computing a linear regression
design_matrix = np.vstack ((np.ones_like(X), X)).T
pinv_dmat = np.linalg.pinv(design_matrix)
theta = pinv_dmat @ Y
theta
```

```python
# Plotting the linear regression
x = np.vstack((np.ones(5), np.linspace(-10, 0, 5))).T
y = x @ theta.T
plt.plot (x[:,1], y, c='red')
# with original points
plt.scatter(X,Y)
```
